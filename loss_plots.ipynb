# Plots of loss values in training

```python
import pickle
import matplotlib.pyplot as plt
import numpy as np

%matplotlib notebook
```

```python
history1 = pickle.load(open('result_ft_1606288602/History.pkl', "rb"))
history2 = pickle.load(open('result_ft_1610367498/History.pkl', "rb"))
```

```python
history1.keys()
```

```python
fig, ax1 = plt.subplots(figsize=(8, 4))

color_1 = 'tab:red'
color_2 = 'tab:blue'
ax1.set_xlabel('Epoch', labelpad=15, fontsize=15)
ax1.set_xticks(range(len(history1['loss'])))
#ax1.set_xticklabels(xlabels)
ax1.set_ylabel('total loss & softmax loss', labelpad=15, fontsize=15)
#ax1.set_ylim(0, 5000)
#ax.set_yticks(np.arange(0, hist.max()+1))
l1 = ax1.plot(history1['loss'], color=color_1)
l2 = ax1.plot(history1['main_out_loss'], color=color_2)

ax2 = ax1.twinx()
color_3 = 'tab:green'
ax2.set_ylabel('center loss', labelpad=15, fontsize=15)
#ax2.set_ylim(0, 50)
l3 = ax2.plot(history1['centerlosslayer_loss'], color=color_3)

leg = l1 + l2 + l3
labs = ["total loss", "softmax loss", "center loss"]
ax1.legend(leg, labs, loc=1, fontsize=11)

#fig.tight_layout()
plt.title('Fine Tuning with center loss lambda = 0.01')
plt.show()
```

```python
plt.savefig('plotloss_funetuning2.jpg')
```
